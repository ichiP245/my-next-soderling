```python
import pandas as pd
import time
```

```python
pd.set_option('display.max_columns', None)
```

```python
# Dallas tiene desde 2022
torneos = ['frenchopen', 'usopen', 'ausopen', 'wimbledon',        #Grand Slams
  'cincinnati', 'indianwells', 'madrid', 'miami', 'montecarlo', 'montreal', 'paris', 'rome', 'shanghai',   #Master 1000
  'dallas', 'rotterdam', 'doha', 'riodejaneiro', 'acapulco', 'dubai', 'barcelona', 'munich', 'hamburg', 'halle', 'washington', 'tokyo', 'beijing', 'vienna', 'basel' #ATP 500
           ]
```

La extraccion de los csvs tardo un total de 6 minutos

```python
lista_dfs = []
for torneo in torneos:
  for anio in range(2015, 2026):
    try:
      df = pd.read_csv(f'http://www.tennis-data.co.uk/{anio}/{torneo}.csv')
    except BaseException as e:
      print(f'Error: torneo {torneo}, anio {anio} NO registrado')
    else:
      lista_dfs.append(df)
    time.sleep(1)
```

Tenemos 275 torneos

```python
len(lista_dfs)
```

El df en el indice 237 no tiene la columna 'Location'

```python
i = 0
for j in range(len(lista_dfs)):
  if 'Location' not in lista_dfs[j].columns.to_list():
    i+=1
    print(j)
```

Descartamos ese torneo porque se leyo mal

```python
lista_dfs.pop(237)
```

Nos quedan 274 torneos

```python
len(lista_dfs)
```

Buscamos las columnas que estan en todos los dfs

```python
# Definimos como columnas inicales a las del primer df
cols_inic = lista_dfs[0].columns.to_list()
cantidad_cols = [] # Vamos a guardar la cantidad de columnas en cada df

for df in lista_dfs:
  cantidad_cols.append(len(df.columns)) # Guardamos la cantidad de columnas
  for col in cols_inic:
    if col not in df.columns: # Si alguna de las columnas iniciales no esta en un df, la sacamos
      cols_inic.remove(col)
      print(f'Se saco la variable {col} por el df {df['Location'].unique()}')
```

Miramos la cantidad de columnas en cada df. La mayoria tiene 36 columnas, que son con las que nos quedamos

```python
from collections import Counter
Counter(cantidad_cols)
```

Cerramos con 36 columnas

```python
len(cols_inic)
```

Armamos una nueva lista de dfs, con solo esas columnas, y para que quede ese orden

```python
new_lista_dfs = []
for df in lista_dfs:
  new_lista_dfs.append(df[cols_inic])
```

Nos aseguramos de tener los 274 dfs

```python
len(new_lista_dfs)
```

Los concatenamos

```python
df = pd.concat(new_lista_dfs)
```

Guardamos eso como CSV

```python
df.to_csv('TenisCSV.csv')
```
